```python
cat_var=['Gender','Education Level','Job Title']
```

```python
for var in cat_var:
    df[var]=df[var].fillna(df[var].mode()[0])
```

- The code is a loop that iterates over each categorical variable in the list (cat_var). 
- For each variable, it fills missing values with the most frequent value (mode) in that variable. 
- This is a common strategy for handling missing values in categorical data.

```python
num_var=['Age','Years of Experience','Salary']
```

```python
for var in num_var:
    df[var]=df[var].fillna(df[var].median())
```

- The code is a loop that iterates over each numerical variable in the list (num_var). 
- For each variable, it fills missing values with the median value of that variable. 
- Using the median for imputation is a common strategy for handling missing values in numerical data.

# Pie Plot

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution in the Dataset')
plt.show()
```

- This code takes the gender distribution data from a dataset creates a pie chart using Matplotlib and then displays the chart with labels.
- The pie chart visually represents the distribution of genders in the dataset.

# Hist Plot

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Education Level', bins=range(len(df['Education Level'].unique()) + 1), discrete=True, palette='Set3')
plt.title('Distribution of Employees by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- A histogram plot where 'Education Level' is on the x-axis and the y-axis represents the frequency of each education level.

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Salary'], bins=10, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

- This code creates a histogram of the salary distribution using Seaborn and Matplotlib. 
- The histogram provides insights into the distribution of salary values in the dataset.
- The resulting plot is displayed with appropriate axis labels and a title.

# Box Plot

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Salary', data=df, orient='h', width=0.7)
plt.title('Salary Distribution Box Plot')
plt.xlabel('Salary')
plt.show()
```

- This code creates a horizontal box plot of the salary distribution using Seaborn and Matplotlib. 
- A box plot provides a visual summary of the distribution of a dataset.
- The resulting plot is displayed with an appropriate title, x-axis label, and figure size.

# Scatter Plot

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Years of Experience', y='Salary', data=df, hue='Gender', palette='Set2', s=100)
plt.title('Scatter Plot: Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(title='Gender')
plt.show()
```

- This code creates a scatter plot to visualize the relationship between "Years of Experience" and "Salary" in a dataset.
- The points are differentiated by gender, and the plot includes a legend, title, and axis labels for better interpretation.

# Line Plot

```python
average_salary_by_experience = df.groupby('Years of Experience')['Salary'].mean().reset_index()
average_salary_by_experience
plt.figure(figsize=(12, 4))
plt.plot(average_salary_by_experience['Years of Experience'], average_salary_by_experience['Salary'], marker='o')
plt.title('Average Salary by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.grid()
plt.show()
```

- This code calculates the average salary for each level of experience and visualizes the trend using a line plot.
- The resulting plot includes a title, axis labels, and a grid for better interpretation.
